```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
#let's see what this dataset is 
data = sm.datasets.longley.load(as_pandas = False)
print(type(data))
```

```python
data.exog = sm.add_constant(data.exog)
print(type(data))
```

```python
#then print the last 5 arrays (Not sure)
print(data.exog[:5])
```

```python
#let's say the data is heteroskedascticity and we can define the sigam 
ols_resid = sm.OLS(data.endog, data.exog).fit().resid

plt.figure(figsize = (24,10))
plt.plot(ols_resid)
```

```python
#now we want to fit the data
resid_fit = sm.OLS(ols_resid[1:], sm.add_constant(ols_resid[:-1])).fit()

#now we want to t-values
print("t-values:", resid_fit.tvalues[1])

#p-values
print("p-values", resid_fit.pvalues[1])
```
